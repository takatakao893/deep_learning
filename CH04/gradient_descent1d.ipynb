```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

# x軸の値を作成
x_vals = np.arange(-20.0, 20.0, 0.1)
# 2乗のグラフを作成
plt.figure(figsize=(8,6))
plt.plot(x_vals, f(x_vals), label='$y = x^2$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('$f(x) = x^2$', fontsize=20)
plt.grid()
plt.show()
```

```python
# 接点を指定
x = -5

def numerical_diff(f, x):
    h = 1e-4 # 0.0001
    return (f(x+h) - f(x-h)) / (2*h)

# 微分(傾き)を計算
dx = numerical_diff(f, x)
print(dx)

# 切片を計算
b = f(x) - dx * x
print(b)
```

```python
# 接線を作図
plt.figure(figsize=(8, 6))
plt.plot(x_vals, f(x_vals), label='f(x)')
plt.plot(x_vals, dx * x_vals + b, label="f'(x)")
plt.scatter(x, f(x))
plt.scatter(0.0, 0.0, color='red', s=100, marker='x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.suptitle('$f(x) = x^2$', fontsize=20)
plt.title('(x, f(x))=(' + str(x) + ',' + str(np.round(f(x), 2)) + ')' +
          ', dx=' + str(np.round(dx, 2)), loc='left')
plt.legend()
plt.grid()
plt.ylim(-50, 400)
plt.show()
```

```python
# 関数を作成
def f(x):
    return (x + 4)*(x + 1)*(x - 1)*(x - 3)

# 接点を指定
x = 2.222

# 微分(傾き)を計算
dx = numerical_diff(f, x)

# 切片を計算
b = f(x) - dx * x

# x軸の値を生成
x_vals = np.arange(-5.0, 4.0, 0.1)

# 接線を作図
plt.figure(figsize=(8, 6))
plt.plot(x_vals, f(x_vals), label='$y = x^2$')
plt.scatter(x, f(x))
plt.plot(x_vals, dx * x_vals + b)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('dx=' + str(np.round(dx, 3)), loc='left')
plt.grid()
plt.show()
```
